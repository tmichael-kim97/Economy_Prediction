countried hosting olympics before 1990. 

ii) We researched on the net , especially exploring World Tourism Organization dataset, but unable to find 
at each country level prior to 1990.

i) The data was not consistent across the countries.-For Tourism , almost no country had complete data for +-8.

ii) For many indicator variables the data just didn't exist. Most of them had missing values for almost all the observations

**S&W**: Whether the Olympics was a Summer Olympics or a Winter one  
**B&A**: Whether the Year was before 1984 or after 1984

# I) Tourism Prediction based on Economic Indicators

The actual Tourism Arrival values are converted into Log diff  
where **logdiffT=ln(Tourism_thisyear)-ln(Tourism_lastyear)**

```python
# Reading the 8 year window dataset for Tourism 
tourism=pd.read_excel('tourism_data.xlsx', 'Sheet1', index_col=None, na_values=['NA'],header=True)
tourism.head()
```

```python
# Dropping the Original Tourism variable and Year variable
tourism=tourism.drop(['ST.INT.ARVL','year'],axis=1)
tourism.head()
```

Checking for % Non-Missing values in each feature

```python
# Finding the % non Missing in each variable
NMissing=(tourism.describe().ix[0]/len(tourism))*100
NMissing.sort(ascending=False)
NMissing
```

Keeping only the variables which have less than 20% missing values

```python
var_to_drop=NMissing[NMissing.abs()<80].index
tourism=tourism.drop(var_to_drop,axis=1)
print("Total Variables Dropped:",len(var_to_drop))
print("Total Variables left in Dataframe:",tourism.shape[1])
tourism.head()
```

```python
# Re-coding Country variable to nominal
tourism = tourism.ffill()
tourism['country'] = tourism['country'].factorize()[0]
```

```python
tourism.head(3)
```

```python
# Visualizing by plotting scatter plots. 
pd.scatter_matrix(tourism.ix[:,4:8], alpha=0.2, figsize=(10, 10), diagonal='kde')
```

Checking correlation between GDP variables to retain one. 

```python
# Getting the columns indices which are pure GDP variables. 
columns=['NY.GDP.DEFL.ZS_logdiff', 'NY.GDP.FRST.RT.ZS',
'NY.GDP.MINR.RT.ZS', 'NY.GDP.MKTP.CD_logdiff', 'NY.GDP.MKTP.CN_logdiff', 'NY.GDP.MKTP.KD_logdiff',
'NY.GDP.MKTP.KD.ZG', 'NY.GDP.MKTP.KN_logdiff', 'NY.GDP.MKTP.PP.CD_logDiff', 'NY.GDP.MKTP.PP.KD_logdiff', 
'NY.GDP.NGAS.RT.ZS', 'NY.GDP.PCAP.CD_logdiff', 'NY.GDP.PCAP.KD.ZG', 'NY.GDP.PCAP.KN_logdiff', 
'NY.GDP.PCAP.PP.CD_logdff', 'NY.GDP.PCAP.PP.KD_logdiff']
tourism[columns].corr()
# Keeping just the GDP per capita growth (NY.GDP.PCAP.KD.ZG) and only those GDP variables which have less correlation (<0.5)
# with NY.GDP.PCAP.KD.ZG.Removing others from/ dataframe
shortlist=list(tourism[columns].corr().ix[12][tourism[columns].corr().ix[12].abs() <0.5].index)
shortlist.append('NY.GDP.PCAP.KD.ZG') # Appending the GDP growth variable

# Getting the above remaining GDP columns removed from the original dataframe
dropCol=set(columns)-set(shortlist)
tourism=tourism.drop(dropCol,axis=1)  # Dropping the above GDP variables. 
print("Total variables dropped:",len(dropCol))
print("Total variables left:",tourism.shape[1])
```

# Feature Selection: Using Random Forest, Anova & Stability Selection

Our Initial approach for feature selection is to check for top performing features from three feature
selection methods: Random Forests, Univariate Selection (Anova) and Stability Selection (LARS)

```python
test_data=tourism.fillna(0.0000099)# Replacing missing values with -0.0000099
#dropping the Year, country and SW feature to check only on other numeric variable importance importance
test_data=test_data.drop(test_data.columns[[1,2,3]],axis=1) 
tourism_Y=test_data.ix[:,0] # Separating target variable from rest of data
tourism_X=test_data.ix[:,1:]
test_data.head()
```

# i) Random Forest: Mutual Information 

Ranking the top 25 features by importance value and plotting

Our Initial approach to get a gauge of the best set of features/predictors is by using Random Forest

```python
from sklearn import ensemble
est=ensemble.RandomForestRegressor(n_estimators=100,criterion='mse',random_state=0)
est.fit(tourism_X, tourism_Y)
indices=np.argsort(est.feature_importances_)[::-1]
var_rf=indices[:12] # Rf Gave us 12 best features. 
tourism_X.columns[indices[:12]] # Top 12 features
```

```python
# Printing feature ranking and plotting the ranked features
# Print the feature ranking
var=25
top_var=indices[:var]
print("Feature ranking: Mutual Information")
for f in range(var):
    print("%d. feature %d (%f)" % (f + 1, top_var[f], est.feature_importances_[top_var[f]]))

std = np.std([tree.feature_importances_ for tree in est.estimators_],axis=0)
# Plot the feature importances of the forest
plt.figure(figsize=(20,7),dpi=200)
plt.title("Feature importances (Mutual Information)")
plt.bar(range(var),est.feature_importances_[top_var],color="r", yerr=std[top_var], align="center")
plt.xticks(range(var),top_var)
plt.xlim([-1, var-5])
```

# ii) Univariate Selection: F-scores (Anova):

Using F-test (Anova) for feature selection. Ranking the top 25 features by F-score measure and plotting

```python
from sklearn.feature_selection import SelectKBest , f_regression
est_f=SelectKBest(f_regression).fit(tourism_X, tourism_Y)
indices_f=np.argsort(est_f.scores_)[::-1]
var_f=indices_f[:10]
tourism_X.columns[indices_f[:10]]   # Printing the top 10 features. 
```

```python
# Printing feature ranking and plotting the ranked features
# Print the feature ranking
top_var_f=indices_f[:var]
print("Feature ranking: F-scores")
for f in range(var):
    print("%d. feature %d (%f)" % (f + 1, top_var_f[f], est_f.scores_[top_var_f[f]]))
    
# Plot the feature importance using F score measure
plt.figure(figsize=(20,7),dpi=200)
plt.title("Feature importances (F Scores)")
plt.bar(range(var),est_f.scores_[top_var_f],color="r", align="center")
plt.xticks(range(var), top_var_f)
plt.xlim([-1, var-5])
```

# iii) Stability Selection: LARS Stability Measure

Using Randomised Lasso (Least Angle Regression). Ranking the top 25 features by stability score measure and plotting

```python
from sklearn.linear_model import RandomizedLasso
est_rlasso = RandomizedLasso(max_iter=1200)
est_rlasso.fit(tourism_X, tourism_Y)    # Running Randomised Lasso
indices_rl=np.argsort(est_rlasso.scores_)[::-1]
#tourism_X.columns[indices_r]
#shortlistvars=tourism_X.columns[est_rlasso.nonzero]
var_rl= indices_rl[:36]               # Extracting top 15 features. 
tourism_X.columns[indices_rl[:15]]   # LARS gave 15 top features. 
```

```python
# Printing feature ranking and plotting the ranked features
# Print the feature ranking
top_var_rl=indices_rl[:var]
print("Feature ranking: F-scores")
for f in range(var):
    print("%d. feature %d (%f)" % (f + 1, top_var_rl[f], est_rlasso.scores_[top_var_rl[f]]))
    
# Plot the feature importance using F score measure
plt.figure(figsize=(20,7),dpi=200)
plt.title("Feature importances (Stability Scores)")
plt.bar(range(var),est_rlasso.scores_[top_var_rl],color="r", align="center")
plt.xticks(range(var), top_var_rl)
plt.xlim([-1, var-5])
```

```python
print(var_rf)   # Important features from three selection methods. 
print(var_f)
print(var_rl)  
```

Keeping the features which are atleast present in 2 or more sets. 

```python
vars=(set(list(var_rf)) & set(list(var_f))) | (set(list(var_rf)) & set(list(var_rl))) | (set(list(var_f)) & set(list(var_rl)))
```

```python
var_final=list(tourism.columns[[2,3]]) + list(tourism_X.columns[list(vars)]) # Adding first three categorical ones
```

```python
var_final
```

# Running Lasso on the combined common features and Target variable

```python
tourism_train=tourism[var_final] 
# Getting the training data with only the said features. 
```

```python
tourism.head()
```

Encoding the categorical variables

```python
# Encoding categorical variables
from sklearn import preprocessing
tourism_train=preprocessing.Imputer(missing_values='NaN', strategy='mean', axis=0).fit_transform(tourism_train)
tourism_train=preprocessing.OneHotEncoder(categorical_features=[1],sparse=False).fit_transform(tourism_train)
```

```python
# Standardising the variables
tourism_train=preprocessing.scale(tourism_train)
#tourism_Y=preprocessing.scale(tourism_Y)
```

```python
#Running Lasso: 
from sklearn import linear_model 
lasso=linear_model.LassoCV(max_iter=2000,cv=10,normalize=False)
lasso.fit(tourism_train,tourism_Y)
lasso.alpha_
scores=np.zeros((100,1))
scores[:,0]=np.mean(lasso.mse_path_,axis=1)
scores=np.sort(scores)
lasso.coef_
```

Getting the final coefficient along with the features. 

```python
# Creating a table of feature weights
features=['Intercept']+[var_final[0]]+['B&A_Before']+['B&A_During']+['B&A_After']+var_final[2:]
weights=[lasso.intercept_] +[lasso.coef_[0]]+[lasso.coef_[1]]+[lasso.coef_[2]]+[lasso.coef_[3]]+list(lasso.coef_[4:])
coef=pd.Series(weights,index=features)
print(" The feature-weights matrix is:")
coef[coef.abs()>0]
```
